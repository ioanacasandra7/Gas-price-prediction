```python
#dependencies
import pandas as pd
import requests
from pprint import pprint as prettyprint
from config import api_key

#get urls for the gas price regions
east_url = f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.EMM_EPM0_PTE_R10_DPG.W'
mid_url=f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.EMM_EPM0_PTE_R20_DPG.W'
gulf_url=f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.EMM_EPM0_PTE_R30_DPG.W'
rock_url=f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.EMM_EPM0_PTE_R40_DPG.W'
west_url=f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.EMM_EPM0_PTE_R50_DPG.W'

gas_urls=[east_url,mid_url,gulf_url,rock_url,west_url]
region_labels=['east','midwest','gulf','rocky','west']
```

```python
gas_dfs=[]
i=0
for url in gas_urls:
    
    # get the price data for the region
    response = requests.get(url).json()
    
    gas_prices_df=pd.DataFrame(response['series'][0]['data'])
    
    # rename the columns and convert the date column to a datetime data type.
    new_df=gas_prices_df.rename(columns={0:'date',1:f'{region_labels[i]}_price'})
    new_df['date']= pd.to_datetime(new_df['date'])
    new_df['date'] = new_df.date + pd.Timedelta(days=4)
    gas_df=new_df.sort_values(['date'])

    gas_dfs.append(gas_df)
    
    i+=1
    
gas_dfs[0]
```

```python
#get urls for the U.S. Exports of Crude Oil and Petroleum Products, Weekly (Thousand Barrels per Day)
exports_url = f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.WTTEXUS2.W'

response = requests.get(exports_url).json()
export_prices=response['series'][0]['data']

#put the retrieved data into a dataframe
export_prices_df=pd.DataFrame(export_prices)

# rename the columns and convert the date column to a datetime data type.
export_df=export_prices_df.rename(columns={0:'date',1:'crude_exp_price'})
export_df['date']= pd.to_datetime(export_df['date'])
export_df=export_df.sort_values(['date'])
export_df
```

```python
# Weeky imports for crude oil data
imports_url=f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.WTTIMUS2.W'

response=requests.get(imports_url).json()
import_prices=response['series'][0]['data']

#put the retrieved data into a dataframe
import_prices_df=pd.DataFrame(import_prices)
import_prices_df

# rename the columns and convert the date column to a datetime data type.
import_df=import_prices_df.rename(columns={0:'date',1:'crude_imp_price'})
import_df['date']= pd.to_datetime(import_df['date'])
import_df=import_df.sort_values(['date'])
import_df
```

```python
# Weeky imports for crude oil data
wti_url=f'http://api.eia.gov/series/?api_key={api_key}&series_id=PET.RWTC.W'

response=requests.get(imports_url).json()
# prettyprint(response)
wti_prices=response['series'][0]['data']
wti_prices
# #put the retrieved data into a dataframe
wti_prices_df=pd.DataFrame(wti_prices)
wti_prices_df

# rename the columns and convert the date column to a datetime data type.
wti_df=wti_prices_df.rename(columns={0:'date',1:'wti_price'})
wti_df['date']= pd.to_datetime(wti_df['date'])
wti_df=wti_df.sort_values(['date'])
wti_df
```

```python
# #create a final csv
final_df=import_df.merge(export_df,on='date')\
                .merge(gas_dfs[0],on='date')\
                .merge(gas_dfs[1],on='date')\
                .merge(gas_dfs[2],on='date')\
                .merge(gas_dfs[3],on='date')\
                .merge(gas_dfs[4],on='date')\
                .merge(wti_df,on='date')\
                .sort_values(['date'])
final_df
```

```python
final_df.set_index('date').to_csv('../Data/gas_oil.csv')
```
